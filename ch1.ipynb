```python
# problem 1.1 -
def luminosity(age):
    """Get luminosity of the Sun at any age, relative to the luminosity today.

    Parameters
    ==========
    age : float
        Time, in years, after the Sun's creation.

    Returns
    =======
    lum : float
        The ratio of the Sun's luminosity then to the Sun's luminosity now. Assumes
        the Sun is still on the Main sequence.
    """
    current_age = 4.6e9
    return (1 / (1 + 0.4 * (1 - age/current_age)))

def problem1():
    s = pd.Series([luminosity(t * 1e6) for t in range(5000)])
    sns.tsplot(s)

problem1()
```

```python
# problem 3 - 
def dry_adiabat(pressures):
    ground_temp = 300
    surface_pressure = 1000
    return ground_temp * (pressures/surface_pressure) ** (2/7)

def K_to_F(kelvins):
    return (kelvins - 273) * 1.8 + 32

def problem3a():
    pressures = np.arange(10, 1000)
    df = pd.DataFrame({"pressures": pressures,
          "dry_adiabat_K": dry_adiabat(pressures),
          "dry_adiabat_F": K_to_F(dry_adiabat(pressures))})
    sns.regplot("pressures", "dry_adiabat_K", df, fit_reg=False)
    sns.regplot("pressures", "dry_adiabat_F", df, fit_reg=False)
problem3a()
```
