```python
# problem 1.1 -
def luminosity(age):
    """Get luminosity of the Sun at any age, relative to the luminosity today.

    Parameters
    ==========
    age : float
        Time, in years, after the Sun's creation.

    Returns
    =======
    lum : float
        The ratio of the Sun's luminosity then to the Sun's luminosity now. Assumes
        the Sun is still on the Main sequence.
    """
    current_age = 4.6e9
    return (1 / (1 + 0.4 * (1 - age/current_age)))

def problem1():
    s = pd.Series([luminosity(t * 1e6) for t in range(5000)])
    plt.figure(figsize=(12,9))
    plt.plot(s)

problem1()
```

```python
def func(x):
    return x - e**(-x)


def sign(x):
    try:
        return x/abs(x)
    except ZeroDivisionError:
        return 0
       

def bisect(interval, fn, iters):
    xmin = interval[0]
    xmax = interval[1]
    xmid = sum(interval)/2
    if xmax - xmin < 1e-8:
        return (xmid, iters)
    low, mid, hi = [fn(x) for x in [xmin, xmid, xmax]]
    if sign(low) == sign(hi):
        return None
    if sign(low) == sign(mid):
        return bisect((xmid, xmax), fn, iters+1)
    else:
        return bisect((xmin, xmid), fn, iters+1)


def newton(x0, fn, iters):
    x1 = x0 - fn(x0)/derivative(fn, x0, dx=1e-5)
    if abs(x1 - x0) < 1e-8:
        return ((x1 + x0)/2, iters)
    else:
        return newton(x1, fn, iters+1)
            
        
def problem5():
    print(bisect((0, 1), func, 0))
    print(newton(0.5, func, 0))

problem5()
```

```python
def gaussian(x):
    return e ** (-x**2)


def problem6():
    plt.figure(figsize=(12,9))
    aspace = np.linspace(0, 10, 200)
    plt.title("Romberg integration of $y=\int_0^ae^{-x^2}dx$\n", fontsize=24)
    plt.ylabel("$y$", fontsize=24)
    plt.plot(aspace, [problem6b(a) for a in aspace])
    plt.text(2, 0.1, r"number of steps = $2^{\lceil log_2 a\rceil}\cdot32+1$", fontsize=18)
    plt.xlabel("$a$", fontsize=24)


def problem6b(a):
    if a == 0:
        return 0
    xspace = np.linspace(0, a, 2**(5 + np.ceil(log2(a)))+1)
    return romb(gaussian(xspace), xspace[1] -xspace[0])

problem6()
```

```python
def diff(a):
    return (lambda y, t: a*y*(1-y))

def problem7():
    tscale = np.linspace(0, 10, 100)
    diffeq = odeint(diff(2), 0.01, tscale)
    plt.figure(figsize=(12, 8))
    plt.text(3.6, 1.1, r"$\frac{dY}{dt} = aY\cdot(1-Y)$", fontsize=24)
    plt.plot(tscale, odeint(diff(1), 0.01, tscale))
    plt.text(5, 0.5, r"$a=1$", fontsize=16)
    plt.plot(tscale, odeint(diff(2), 0.01, tscale))
    plt.text(2.5, 0.5, r"$a=2$", fontsize=16)
    plt.plot(tscale, odeint(diff(3), 0.01, tscale))
    plt.text(1.65, 0.5, r"$a=3$", fontsize=16)
    plt.plot(tscale, odeint(diff(4), 0.01, tscale))
    plt.text(0.6, 0.5, r"$a=4$", fontsize=16)
    plt.xlabel("time (t)")

problem7()
```

```python
# "Inspired" by http://wiki.scipy.org/Cookbook/Zombie_Apocalypse_ODEINT
def func8(xs, t):
    y = xs[0]
    z = xs[1]
    fy = - z/(y**2 + z**2)
    fz = y/(y**2 + z**2)
    return [fy, fz]
    
def problem8():
    y0 = 1
    z0 = 0
    tscale = np.linspace(0, 10, 100)
    sol = odeint(func8, [y0, z0], tscale).transpose()
    plt.figure(figsize=(12, 8))
    plt.plot(tscale, sol[0])
    plt.text(1.7, 0, r"$Y$", fontsize=24)
    plt.text(0.1, -0.5, r"$dY/dt = -\frac{Z}{Y^2+Z^2}$", fontsize=18)
    plt.plot(tscale, sol[1])
    plt.text(3.4, 0, r"$Z$", fontsize=24)
    plt.text(2.8, 0.5, r"$dZ/dt = \frac{Y}{Y^2+Z^2}$", fontsize=18)
    plt.xlabel("time (t)")

problem8()
```
